### Age of people who left the company

```python
age_for_yes = []
for age, group in df[df['Attrition'] == 'Yes'].groupby('Age'):
    age_for_yes.append(group.shape[0])
```

### Age of people who didnt left the company

```python
age_for_no = []
for age, group in df[df['Attrition'] == 'No'].groupby('Age'):
    age_for_no.append(group.shape[0])
```

### Age of employees

```python
Age = []
for age in df['Age'].unique():
    Age.append(age)
```

```python
len(Age),len(age_for_yes)
```

### Plot the two charts 

```python
plt.plot(Age[:len(age_for_yes)], age_for_yes, marker='o',label='yes')
plt.plot(Age[:len(age_for_no)], age_for_no, marker='o',label='no')
plt.xlabel('Age')
plt.ylabel('Count of Employees')
plt.title('Age vs Attrition ')
plt.legend()
plt.show()
```

```python
for column in df.columns:
    for col, group in df[df['Attrition'] == 'Yes'].groupby(column):
        print(col,group.shape[0])
```

### Plot the attrition for all the columns

```python
for column in df.columns:
    age_for_yes = []
    age_for_no = []
    column_values_0 = []
    
    column_values_1 = []
    
    for value, group in df[df['Attrition'] == 'Yes'].groupby(column):
        age_for_yes.append(group.shape[0])
        column_values_0.append(value)
    
    
    for values,groups in df[df['Attrition'] == 'No'].groupby(column):
            age_for_no.append(groups.shape[0])
            column_values_1.append(values)
    plt.figure(figsize=(10,6))
    plt.plot(column_values_0, age_for_yes, marker='o', label='yes')
    plt.plot(column_values_1, age_for_no, marker='o', label='no')
    plt.xlabel(column)
    plt.ylabel('Count of Employees')
    plt.title(f'{column} vs Attrition')
    plt.legend()
    plt.show()
```
